```python
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import shutil
import datetime

import os
import sys
sys.path.append(os.path.dirname(os.path.abspath(''))+'/wfdb-python')
import wfdb
from wfdb import processing
```

```python
#PATH_DATASET = '/home/nsc/seers/dataset'
PATH_DATASET = '/home/nsc/seers/temp'
MODEL_NAME = 'life-threatening_arrhythmias'

ARRHYTHMIAS_NAME_1 = ['Asystole', 
                      'Bradycardia', 
                      'Tachycardia', 
                      'Ventricular_Tachycardia', 
                      'Ventricular_Flutter_Fib'
                     ]
```

```python
#df = pd.DataFrame(columns=['file', 'label', 'arrhythmia'])
dic_samples = {}
labels = os.listdir(os.path.join(PATH_DATASET, MODEL_NAME))
for label in labels:
    dic_arrhythmia = {}
    arrhythmias = os.listdir(os.path.join(PATH_DATASET, MODEL_NAME, label))
    for arrhythmia in arrhythmias:
        directories = os.listdir(os.path.join(PATH_DATASET, MODEL_NAME, label, arrhythmia))
        list_dir = []
        for directory in directories:
            if os.listdir(os.path.join(PATH_DATASET, MODEL_NAME, label, arrhythmia, directory)):
                list_dir.append(directory)                
                # print (os.path.join(PATH_DATASET, MODEL_NAME, label, arrhythmia, directory, directory+'.csv'),
                #        label, 
                #        arrhythmia)
                # df = df.append({'file':os.path.join(PATH_DATASET, MODEL_NAME, label, arrhythmia, directory, directory+'.csv'), 
                #              'label':label, 'arrhythmia':arrhythmia}, ignore_index=True)
                
        dic_arrhythmia[arrhythmia] = list_dir
    dic_samples[label] = dic_arrhythmia
#df.to_csv("dataset.csv", index = False, mode='w')
```

```python
label = 'true'
#label = 'false'
#arrhythmia = 'Asystole'
#arrhythmia = 'Bradycardia'
#arrhythmia = 'Tachycardia'
arrhythmia = 'Ventricular_Tachycardia'
#arrhythmia = 'Ventricular_Flutter_Fib'
count = len(dic_samples[label][arrhythmia])
print ('record count : ', count)
for idx, record_name in enumerate(dic_samples[label][arrhythmia][:10]):
    #record_name = 'a608s-a1'
    path = os.path.join(PATH_DATASET, MODEL_NAME, label, arrhythmia, record_name)
    record_sample = wfdb.rdrecord(os.path.join(path, record_name))
    str_title = ("%d/%d - %s - %s - %s" % 
                      (idx+1, count,
                       record_sample.record_name,
                       record_sample.comments[0],
                       record_sample.comments[1])
                    )
    print(str_title)
    wfdb.plot_wfdb(record=record_sample, time_units='seconds', 
                   title=str_title, figsize=(10, 4))
           
    #plt_fig = wfdb.figure_wfdb(record=record_sample, time_units='seconds', 
    #                   title=str_title, figsize=(10, 4), return_fig=True)
    #plt_fig.savefig(record_name,dpi=100)
    
    display(record_sample.__dict__)
    if np.all(np.isnan(record_sample.p_signal)):
        display(record_sample.__dict__)
    if np.all(record_sample.p_signal == 0.):
        display(record_sample.__dict__)
    #break
```
