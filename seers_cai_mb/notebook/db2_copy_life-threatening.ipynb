```python
def arrhythmias_name(comment):
    for name in ARRHYTHMIAS_NAME_2:
        if name in comment:
            return ARRHYTHMIAS_NAME_2TO1[name]
    return None
```

```python
with open(os.path.join(PATH_MIMIC2, FILE_RECORD_LIST)) as f:
    for line in f:
        list_record.extend(line.splitlines())
```

```python
print('target record number : ', len(list_record))
for num, record in enumerate(list_record):
    #if num < 7 : continue
    source = os.path.join(PATH_MIMIC2, record)
    target = os.path.join(PATH_MIMIC2_LIFE_THREATENING, record)
    print (num, 'source : ', source)
    print (num, 'target : ', target)
    if os.path.isdir(target):
        print ('exist directory.')
        continue
    shutil.copytree(source, target)
    
```
