```python
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import shutil
import datetime

import os
import sys
sys.path.append(os.path.dirname(os.path.abspath(''))+'/wfdb-python')
import wfdb
from wfdb import processing
sys.path.append(os.path.dirname(os.path.abspath(''))+'/mc100')
from mc100_class import MC100
```

```python
PATH_MC100 = '/home/nsc/seers/cardio/database/mc100/'
MC100_CSV = 'seers_cardio_database_mc100.csv'
MC100_CSV = os.path.join(PATH_MC100, MC100_CSV)

dic_dtype = {'measurementId': 'i',
            'measurementCode': 'S',
            'measurementStatus': 'i',
            'registerUserId': 'i',
            'organizationId': 'i',
            'patientCode': 'S',
            'startDateTime': 'S',
            'endDateTime': 'S',
            'downloadDateTime': 'S',
            'ECGUrl': 'S',
            'fileSize': 'i',
            'sampling': 'i',
            'gain': 'f',
            'endian': 'i',
            'analysisUrl': 'S',
            'storePath': 'S',
            'pridictStatus': 'S',
            'makeDatasetStatus': 'S'}

dic_schema = {'measurementId':np.array([], dtype='i'),
            'measurementCode': np.array([], dtype='S'),
            'measurementStatus': np.array([], dtype='i'),
            'registerUserId': np.array([], dtype='i'),
            'organizationId': np.array([], dtype='i'),
            'patientCode': np.array([], dtype='S'),
            'startDateTime': np.array([], dtype='S'),
            'endDateTime': np.array([], dtype='S'),
            'downloadDateTime': np.array([], dtype='S'),
            'ECGUrl': np.array([], dtype='S'),
            'fileSize': np.array([], dtype='i'),
            'sampling': np.array([], dtype='i'),
            'gain': np.array([], dtype='f'),
            'endian': np.array([], dtype='i'),
            'analysisUrl':np.array([], dtype='S'),
            'storePath': np.array([], dtype='S'),
            'pridictStatus': np.array([], dtype='S'),
            'makeDatasetStatus': np.array([], dtype='S')}

if os.path.isfile(MC100_CSV):
    df_mc100 = pd.read_csv(MC100_CSV, sep=',', dtype=dic_dtype)
else:
    df_mc100 = pd.DataFrame(dic_schema)

print (df_mc100.dtypes)   
df_mc100
```

```python
# load MC100 raw data

"""
measurementStatus
0x00 : NONE
0x01 : RECODING_START
0x02 : RECODING_END
0x03 : UPLOAD_COMPLETE
0x04 : ANALYZING
0x05 : READY
0x06 : REVIEWING
0x07 : COMPLETED
0x0A : REPORT RELEASE
0x0B : RE ANALYZING
0x0F : ERROR
"""
mc100 = MC100('/home/nsc/seers/cardio/database/mc100') #여기 위치 어디 ? 
DOWNLOAD_PATH = mc100.DOWNLOAD_PATH
print ('Load Path : ', DOWNLOAD_PATH)

# get list target load
# mc100.get_list([0x05, 0x06])
mc100.get_list([0x05, 0x06, 0x07]) # for test

for idx, measurement in enumerate(mc100.list_measurement):    
    if measurement['ecgDataList'] is None:        
        continue
    if measurement['analysisOutputFileList'] is None:        
        continue
        
    if df_mc100.loc[df_mc100['measurementId'] == measurement['measurementId']].index.size is not 0 :
        print ('existed measurementId - index:',
               df_mc100.loc[df_mc100['measurementId'] == measurement['measurementId']].index)
        continue
   
    # download ecg, annotation
    print ('download_ecg - ', measurement['measurementCode'])
    mc100.download_ecg([measurement])

    # create record header
    print ('create record header - ', measurement['measurementCode'])
    dt = datetime.datetime.strptime(measurement['ecgDataList'][0]['startDateTime'],'%Y-%m-%d %H:%M:%S')
    path = os.path.join(DOWNLOAD_PATH, measurement['measurementCode'])
    record = wfdb.Record(record_name=measurement['measurementCode'], 
                        n_sig=measurement['ecgDataList'][0]['sequenceNumber'], 
                        fs=measurement['ecgDataList'][0]['sampling'], 
                        sig_len=int(measurement['ecgDataList'][0]['fileSize'] / 2), 
                        base_time=dt.time(), 
                        base_date=dt.date(),
                        file_name=[measurement['ecgDataList'][0]['filename']],
                        fmt=['16'],
                        adc_gain=[measurement['ecgDataList'][0]['gain']],
                        baseline=[0],
                        units=['mV'],
                        adc_res=[16],
                        adc_zero=[0],
                        init_value=[0],
                        checksum =[0],
                        block_size=[0],
                        sig_name=['II'])
    display(record.__dict__)
    record.wrheader(write_dir=path)
           
    # download annotation
    print ('download_annotation - ', measurement['measurementCode'])
    mc100.download_annotation([measurement])

    # create beat annotation file
    record_name = measurement['measurementCode']
    print ('create beat annoation - ', record_name)
    path = os.path.join(DOWNLOAD_PATH, record_name)
    analysis_file = measurement['analysisOutputFileList'][0]['filename']
    print (path, analysis_file)
    
    df = pd.read_csv(os.path.join(path, analysis_file), sep=' ', header=None)
    df.columns = ['second', 'class', 'classNumber', 'beatNumber']
    
    fs = measurement['ecgDataList'][0]['sampling']
    sample = df['second'].to_numpy()
    sample = sample * fs
    sample = sample.astype('i')
    
    symbol = df['class'].tolist()
    aux_note = df['classNumber'].tolist()
    ann = wfdb.Annotation(record_name=record_name, extension='atr',
                          sample=sample, symbol=symbol, aux_note=aux_note, fs=fs)
    ann.wrann(write_fs=True, write_dir=path)
    
    dic_record = {'measurementId':measurement['measurementId'],
                'measurementCode': measurement['measurementCode'],
                'measurementStatus': 0x05, 
                'registerUserId': measurement['registerUserId'], 
                'organizationId': measurement['organizationId'], 
                'patientCode': measurement['patientCode'], 
                'startDateTime': measurement['startDateTime'], 
                'endDateTime': measurement['endDateTime'], 
                'downloadDateTime': datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'), 
                'ECGUrl': measurement['ecgDataList'][0]['ECGUrl'],
                'fileSize': measurement['ecgDataList'][0]['fileSize'],
                'sampling': measurement['ecgDataList'][0]['sampling'],
                'gain': measurement['ecgDataList'][0]['gain'],
                'endian': measurement['ecgDataList'][0]['endian'],
                'analysisUrl': measurement['analysisOutputFileList'][0]['analysisUrl'],
                'storePath': measurement['storePath'],
                'pridictStatus': False,
                'makeDatasetStatus': False}
    df_mc100 = df_mc100.append(dic_record, ignore_index=True)
    
if len(df_mc100) is not 0:
    df_mc100.to_csv(MC100_CSV, index = False, mode='w')
```

```python
# load MC100 review data

"""
measurementStatus
0x00 : NONE
0x01 : RECODING_START
0x02 : RECODING_END
0x03 : UPLOAD_COMPLETE
0x04 : ANALYZING
0x05 : READY
0x06 : REVIEWING
0x07 : COMPLETED
0x0A : REPORT RELEASE
0x0B : RE ANALYZING
0x0F : ERROR
"""
mc100 = MC100('/home/nsc/seers/cardio/database/mc100')
DOWNLOAD_PATH = mc100.DOWNLOAD_PATH
print ('Load Path : ', DOWNLOAD_PATH)

# get list review load
#mc100.get_list([0x07, 0x0A])
mc100.get_list([0x06, 0x07, 0x0A]) # for test

# create alarm annotation file
for idx, measurement in enumerate(mc100.list_measurement):       
    df_mc100_measurement_index = df_mc100.loc[df_mc100['measurementId'] == measurement['measurementId']].index
    print ('df_mc100 measurement index : ', df_mc100_measurement_index)
    
    if df_mc100_measurement_index.size is 0:
        print ('skip : not stored measurement - ', measurement['measurementCode'])
        continue

    print ('df_mc100 measurement status : ', df_mc100['measurementStatus'][df_mc100_measurement_index[0]])
    if df_mc100['measurementStatus'][df_mc100_measurement_index[0]] == 0x07:
        print ('skip : aleady stored review')
        continue
        
    # get review
    print ('get review - ', measurement['measurementCode'])
    mc100.get_review([measurement])

    if 'analysisBeatClassifySimpleList' not in measurement.keys():
        continue
   
    if measurement['analysisBeatClassifySimpleList'] is None:        
        continue
        
    record_name = measurement['measurementCode']
    print ('create alarm annoation - ', record_name)
    
    path = os.path.join(DOWNLOAD_PATH, record_name)
    fs = measurement['ecgDataList'][0]['sampling']
    sample = np.array([], dtype=int)  
    chan = np.array([], dtype=int)  
    symbol = []
    aux_note = []
    
    for analysis_beat in measurement['analysisBeatClassifySimpleList']:
        if analysis_beat['beatClass'] in ['VT']:
            sample = np.append(sample, analysis_beat['seconds'] * fs)
            symbol.append('"')
            aux_note.append('*** V-TACH ')
            chan = np.append(chan, 1) # true : 1, false = 3
        if analysis_beat['beatClass'] in ['VFL']:
            sample = np.append(sample, analysis_beat['seconds'] * fs)
            symbol.append('"')
            aux_note.append('*** V-FIB/TACH ')
            chan = np.append(chan, 1) # true : 1, false = 3
    
    sample = sample.astype('i')
    if len(sample) > 0:
        ann_alarm = wfdb.Annotation(record_name=record_name, extension='alM',
                                      sample=sample, symbol=symbol, aux_note=aux_note, 
                                      chan=chan, fs=fs)
        ann_alarm.wrann(write_fs=True, write_dir=path)
        # display(ann_alarm.__dict__)
    
    df_mc100['measurementStatus'][df_mc100_measurement_index[0]] = 0x07

if len(df_mc100) is not 0:
    df_mc100.to_csv(MC100_CSV, index = False, mode='w')
```

```python
# MC100 data view

count = len(mc100.list_measurement)
for idx, measurement in enumerate(mc100.list_measurement):
    if measurement['ecgDataList'] is None:
        continue
    
    record_name = measurement['measurementCode']
    fs = measurement['ecgDataList'][0]['sampling']
        
    path = os.path.join(DOWNLOAD_PATH, record_name)
    #start = fs*12
    start = 0
    #end = fs*40
    end = int(measurement['ecgDataList'][0]['fileSize'] / 2) # the end
    annotation = wfdb.rdann(os.path.join(path, record_name), 'atr',
                            sampfrom = start, sampto = end,)
    annotation.sample = annotation.sample - start
    record = wfdb.rdrecord(os.path.join(path, record_name), 
                           sampfrom = start, sampto = end,)
    str_title = ("%d/%d - %s - %s - %s" % 
                      (idx+1, count,
                       record.record_name,
                       'MC100',
                       'beat annotation')
                    )
    print(str_title)
    wfdb.plot_wfdb(record=record, annotation=annotation,
                   time_units='seconds', title=str_title, 
                   plot_sym=True, figsize=(10,3))
    
    
    #start = fs*12
    start = 0
    #end = fs*40
    end = int(measurement['ecgDataList'][0]['fileSize'] / 2) # the end
    alarm = wfdb.rdann(os.path.join(path, record_name), 'alM',
                            sampfrom = start, sampto = end,)
    alarm.sample = alarm.sample - start
    record = wfdb.rdrecord(os.path.join(path, record_name), 
                           sampfrom = start, sampto = end,)
    str_title = ("%d/%d - %s - %s - %s" % 
                      (idx+1, count,
                       record.record_name,
                       'MC100',
                       'Alarm Viewer')
                    )
    print(str_title)
    wfdb.plot_wfdb(record=record, annotation=alarm,
                   time_units='seconds', title=str_title, 
                   figsize=(10,4))
    display(record.__dict__)
    
```
